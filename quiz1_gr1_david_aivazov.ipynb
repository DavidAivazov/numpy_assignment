```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_link = '/content/survey_results_public.csv' #link of csv file, or local adress
df = pd.read_csv(csv_link) #creating data frame
df.head() #printing head of df
```

```python
df_indexed = df.set_index('ConvertedCompYearly') #setting indexing by column Converted Comp Yearly
df_indexed.head() #printing head of dataframe
```

```python
df_indexed.info() #information about columns, which will help us to create a filter
df_indexed.isnull().sum() # checking null-values in data
```

```python
countries = ['Georgia'] #Write countries, which you want to find
age = ['18-24 years old'] #Write age steps, which you want to find (18-24/25-34... years old)
country_filter = df['Country'].isin(countries) #country filter
age_filter = df['Age'].isin(age) #age filter

filtered_df = df[country_filter & age_filter] #filtered df
filtered_df.head() #output
```

```python
sorted_df = df.sort_values(by=['Age','Country'])
sorted_df.head(10)
```

```python
mean_value = df['WorkExp'].mean()
std_deviation = df['WorkExp'].std()
median_value = df['WorkExp'].median()
minimal_value = df['WorkExp'].min()
maximal_value = df['WorkExp'].max()


print(f"Mean value: {mean_value}")
print(f"Standard deviation: {std_deviation}")
print(f"Median: {median_value}")
print(f"Minimum value: {minimal_value}")
print(f"Maximum value: {maximal_value}")
```

```python
plt.subplot(2, 1, 1)
df['Age'].value_counts().plot(kind='bar', color='aquamarine')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
plt.subplot(2, 1, 2)
df['WorkExp'].value_counts().sort_index().plot(kind='line', marker='o', color='turquoise')
plt.title('Experience Distribution')
plt.xlabel('Experience')
plt.ylabel('Count')
```

```python
age_group_exp = df.groupby('Age')['WorkExp'].mean()
age_group_exp.plot(kind='bar', color='indigo')
plt.title('Average Work Experience by Age Group')
plt.xlabel('Age')
plt.ylabel('Average Work Experience')
plt.xticks(rotation=45)  # rotation 45#
plt.grid(axis='y')  # mesh-lines
plt.show()
```
